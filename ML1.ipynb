```python
import numpy as np
# Create an array from 1 to 20
arr = np.arange(1, 21)
print("Array:")
print(arr)
# 2. Find the shape of the array
print(arr.shape)
# 3. Find the maximum and minimum values
print(arr.max())
print(arr.min())
# 4. Calculate the mean and standard deviation
print(arr.mean())
print(arr.std())
# 5. Reshape the array into a 4 x 5 matrix
print("Reshaped Matrix:")
matrix = arr.reshape(4, 5)
print(matrix)
# 6. Display the second row
print("Second Row:")
print(matrix[1])
# 7. Display the third column
print("Third Column:")
print(matrix[:, 2])
```

```python
import pandas as pd
data = {
    "Roll No": [101, 102, 103, 104, 105],
    "Name": ["Ravi", "Priya", "Kiran", "Anu", "Sita"],
    "Marks": [85, 92, 78, 88, 95]
}
df = pd.DataFrame(data)
print("Student Data:")
print(df)

print("\nFirst Three Records:")
print(df.head(3))

print("\nAverage Marks:")
print(df["Marks"].mean())

print("\nStudent with Highest Marks:")
highest = df["Marks"].max()
print(df[df["Marks"] == highest])

print("\nStudent with Lowest Marks:")
print(df[df["Marks"] == df["Marks"].min()])

grade = []

for marks in df["Marks"]:
    if marks >= 90:
        grade.append("A")
    elif marks >= 80:
        grade.append("B")
    else:
        grade.append("C")
df["Grade"] = grade
print("\nDataFrame with Grade:")
print(df)
```

```python
import pandas as pd
data = {
    "Product": ["Laptop", "Mouse", "Keyboard", "Monitor"],
    "Price": [50000, 500, 1200, 10000],
    "Quantity": [10, 50, 30, 15]
}
df = pd.DataFrame(data)
df["Total Value"] = df["Price"] * df["Quantity"]
print("DataFrame:")
print(df)

print("\nProduct with Highest Total Value:")
print(df[df["Total Value"] == df["Total Value"].max()])

print("\nOverall Inventory Value:")
print(df["Total Value"].sum())
```

```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
# Add the Class column
df["Class"] = iris.target
df.columns = ["Sepal_L", "Sepal_W", "Petal_L", "Petal_W", "Class"]
print("Iris DataFrame:")
print(df)

print("\nFirst Five Records:")
print(df.head())

print("\nShape of the DataFrame:")
print(df.shape)

print("\nColumn Names:")
print(df.columns)

print("\nMissing Values:")
print(df.isnull().sum())

print("\nSummary Statistics:")
print(df.describe())

print("\nSamples in Each Class:")
print(df["Class"].value_counts())
```
