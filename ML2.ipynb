```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('students.csv',index_col="Name")
print(df)

print(df.head(5))
print(df.tail(5))
print(df.shape)
print(df.columns)
print(df.dtypes)
print(df.info('columns'))
print(df.info())
```

```python
#TASK 2
print("mean of attendance")
print(df['Attendance'].mean())

print("avg of external marks")
print(df['ExternalMarks'].mean())

print("avg of internal marks")
print(df['InternalMarks'].mean())

print("max of internal ")
print(df['InternalMarks'].max())

print("max of external ")
print(df['ExternalMarks'].max())

print(df['ExternalMarks'].min())
print(df['InternalMarks'].min())

print(df['InternalMarks'].std())

print(df.describe())
```

```python
#TASK 3
df['TotalMarks']= df['ExternalMarks']+ df['InternalMarks']
df['Percentage']=(df['TotalMarks']/100)*100
print(df)
```

```python
#TASK4
print(df[df['Attendance']> 90])
print(df[df['TotalMarks']>90])
print(df[df['Gender']=='F'])
print(df[df['Gender']=='M'])
print(df[df['ExternalMarks']<50])
```

```python
#TASK5
print(df['TotalMarks'].max())

print(df['TotalMarks'].min())
print(df.nlargest(3,'TotalMarks'))
print(df.nsmallest(3,'TotalMarks'))

abo_df=df['TotalMarks'].mean()
print(df[df["ExternalMarks"] > abo_df])
```

```python
#TASK6
group_M=df.groupby('Gender')
print(group_M.mean())
print(group_M['TotalMarks'].mean())
print(group_M['Attendance'].mean())
```

```python
import matplotlib.pyplot as plt
df = df.reset_index()
# 1. Bar Chart - Total Marks of all students
plt.figure(figsize=(8,5))
plt.bar(df["Name"], df["TotalMarks"])
plt.title("Total Marks of Students")
plt.xlabel("Student Name")
plt.ylabel("Total Marks")
plt.show()

# 2. Histogram - Attendance
plt.figure(figsize=(8,5))
plt.hist(df["Attendance"], bins=5)
plt.title("Attendance Distribution")
plt.xlabel("Attendance")
plt.ylabel("Number of Students")
plt.show()

# 3. Pie Chart - Gender Distribution
gender_count = df["Gender"].value_counts()

plt.figure(figsize=(6,6))
plt.pie(gender_count,
        labels=gender_count.index,
        autopct="%1.1f%%",
        startangle=90)
plt.title("Distribution of Male and Female Students")
plt.show()

# 4. Scatter Plot - Attendance vs Total Marks
plt.figure(figsize=(8,5))
plt.scatter(df["Attendance"], df["TotalMarks"])
plt.title("Attendance vs Total Marks")
plt.xlabel("Attendance")
plt.ylabel("Total Marks")
plt.show()
```
